## Information About Variables

We will use the following predictors in the K-NN classification model predict whether or not a player will subscribe to a PLAICraft newsletter:
- Experience Level of Player (character in orginal dataset, but is changed to factor type): `experience`
- Total Hours Played by Player (double): `played_hours`
- Age of Player (double): `age`

Variables that will not be used as a predicitor in the K-NN classification model:
- Gender of Player (character):`gender` since the k-nn model calculates distances of the nearest neighbors to the unknown dataset, gender includes descrete categories that cannot be used in the calculations. While categorical variables could be converted to numeric values, this is not appropriate for gender because there is no meaningful numeric relationship between categories such as Male, Female, or Other.
- Hashed Email of Players (character):`hashedEmail` it does not give us descriptive/charteristic information about the players.
- Name of the Player (character): `name` it does not give us descriptive/charteristic information about the players. 

Response variable for the K-NN classification model:
- Whether or not the user is subscribe to the PLAICraft newsletter (logical in orginal dataset, but is changed to factor type): `subscribe` in the players.csv dataset will be used to calculate the accuracy of our k-nn classification model and is used in the following graphs to display the relationship of those who subscibe vs the predictor varibles.

## Tidying the Data: 
The columns `hashedEmail`, `name` , and `gender` is removed as it is not relevant to our K-NN classsification model. The `Age` column is also renamed to be in all lowercase for consistency between all column names. Additionally, since the observations in subscribe are decrete categories, the column is changed to a factor type allowing smoother use when dispalying colours within graphs.

After wrangling and tidying the data in the `players` object data frame, the data appears data as each variable forms a column, each observation forms a row, and each cell is a single measurment.

## Possible Errors in the Data Set:
- Players do not have 0hrs of playing

```r
#loading the and filtering unwanted data. 
players <- players |>
    select(-hashedEmail, -name, -gender) |> 
    rename(age = Age)|>
    mutate(subscribe = as_factor(subscribe)) |>     
    mutate(subscribe = fct_recode(subscribe, "Yes" = "TRUE", "No" = "FALSE")) 

players
```

```r
options(repr.plot.width = 12, repr.plot.height = 5)

#Graph 1: Distribution of Players in Different Skill Levels and Whether They Are Subscrbed to the PLAICraft Newsletter
experience_distribution <- players |>
    group_by(experience, subscribe)|>
    summarize(number_of_players = n()) |>
    mutate(experience = factor(experience, levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro"))) |>
    ggplot(aes(x = experience, y = number_of_players, fill = subscribe)) +
    geom_bar(stat= "identity",position = "dodge") +
    labs(x = "Experience Level of Player", y = "Number of Players", fill = "Subscribed to Newsletter") +
    ggtitle("Graph 1: Distribution of Players in Different Skill Levels and Whether They Are Subscrbed to the PLAICraft Newsletter") +
    theme(text = element_text(size = 11)) +
    scale_y_continuous(breaks = seq(from = 0, to = 250, by = 10))
experience_distribution
```

```r
#Experince level that is more likely to sign up for the newsletter

yes_experince_level <- players |>
    filter(subscribe == "Yes")|>
    group_by(experience)|>
    summarize(yes_count = n())|>
    select(yes_count)

experince_level <- players |>
    group_by(experience)|>
    summarize(total_count = n())|>
    bind_cols(yes_experince_level) |>
    mutate(percentage_sign_up_for_newsletter = (yes_count/total_count)*100) |>
    select(experience, percentage_sign_up_for_newsletter)
experince_level
```

```r
options(repr.plot.width = 12, repr.plot.height = 5)

#Graph 2: Distribution of Players Across Different Ages and Whether They Are Subscrbed to the PLAICraft Newsletter
age_disribution <- players |>
    ggplot(aes(x = age, fill = subscribe)) +
    geom_histogram(bins=40) +
    facet_grid(cols = vars(subscribe)) +
    labs(x = "Age of Player", y = "Number of Players", fill = "Subscribed to Newsletter") +
    ggtitle("Graph 2: Distribution of Players Across Different Ages and Whether They Are Subscrbed to the PLAICraft Newsletter") +
    theme(text = element_text(size = 11)) +
    scale_x_continuous(breaks = seq(from = 10, to = 57, by = 5)) +
    scale_y_continuous(breaks = seq(from = 0, to = 80, by = 10))
age_disribution

#facet_grid of yes/no
```

```r
options(repr.plot.width = 12, repr.plot.height = 5)

#Graph 3: Distribution of Players With Different Hours Played and Whether They Are Subscrbed to the PLAICraft Newsletter
hours_disribution <- players |>
    ggplot(aes(x = played_hours, fill = subscribe)) +
    geom_histogram(bins=10) +
    facet_grid(cols = vars(subscribe)) +
    labs(x = "Hours Played", y = "Number of Players", fill = "Subscribed to Newsletter") +
    ggtitle("Graph 3: Distribution of Players With Different Hours Played and Whether They Are Subscrbed to the PLAICraft Newsletter") +
    theme(text = element_text(size = 11)) +
    scale_x_continuous(breaks = seq(from = 0, to = 240, by = 20)) +
    scale_y_continuous(breaks = seq(from = 0, to = 250, by = 20))
hours_disribution

#bar plot or facet_grid
```

```r
#Mean Hour Played for Those Subscribed to the PLAICraft Newsletter vs Not Subscribed

mean_hours <- players |>
    group_by(subscribe) |>
    summarize(mean_hours_played = mean(played_hours))
mean_hours

#Miniumn Hours Played for Those Subscribed to the PLAICraft Newsletter vs Not Subscribed
min_hours <- players |>
    group_by(subscribe) |>
    summarize(min_hours_played = min(played_hours))
min_hours

#Maximum Hours Played for Those Subscribed to the PLAICraft Newsletter vs Not Subscribed
max_hours <- players |>
    group_by(subscribe) |>
    summarize(max_hours_played = max(played_hours))
max_hours
```
